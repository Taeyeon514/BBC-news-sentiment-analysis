```python
import pandas as pd
bbc_news = pd.read_csv('/content/drive/MyDrive/09_transformer/data/bbc_news.csv')
bbc_news
```

```python
bbc_news.info()
```

```python
bbc_news = bbc_news.drop(['pubDate', 'guid', 'link'], axis=1)
bbc_news
```

```python
print(bbc_news['title'][0])
print(bbc_news['description'][0])
```

```python
from transformers import pipeline
import pandas as pd

# 텍스트 결합(선택) – title + description
bbc_news['text'] = (bbc_news['title'].fillna('') + ' ' +
                    bbc_news['description'].fillna('')).str.strip()



results = sentiment_clf(bbc_news['text'].tolist())

# 레이블 정리
def to_int_label(lbl):
    # 다양한 표기 보호: 'LABEL_0', '0', 'negative' 등
    l = str(lbl).lower()
    if l in ("label_0", "0", "negative"): return 0
    if l in ("label_1", "1", "neutral"):  return 1
    return 2  # positive

bbc_news['pred_label'] = [to_int_label(r['label']) for r in results]
bbc_news['pred_score'] = [r['score'] for r in results]

# 문자열 라벨을 보기 좋게 만들어주기
id2str = {0: "negative", 1: "neutral", 2: "positive"}
bbc_news['pred_label_str'] = bbc_news['pred_label'].map(id2str)

# 결과 확인
bbc_news_sentiment = bbc_news[['title','pred_label_str','pred_score']]
```
